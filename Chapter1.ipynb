**Chapter1では基本的なnumpyとpandasの処理を行います**

* sin波形を作成します

* 作成にはnumpyを使用します

* グラフ表示にはmatplotlibとplotlyを用います

* plotlyを用いるとインタラクティブなグラフが作成できます

* pandasのdataframe形式でもグラフ表示ができます

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
freq = 250    #周波数
starttime = 0 #開始時刻
endtime  = 10 #終了時刻
t = np.linspace(starttime, endtime, freq) #0~10秒までを250Hz刻みで作成します
print(t[:10]) #始めの5つの要素を確認しましょう
```

```python
y = np.sin(t)
print(y[:10]) #sin(t)も始めの5つの要素を確認しましょう
```

```python
plt.plot(t,y)
```

```python
fig = px.line(x=t, y=y)
fig.show()
```

```python
df = pd.DataFrame(data=t, columns=['time'], dtype='float') #pandasのdataframe形式にしてみましょう
df
```

```python
df['sin_t'] = np.sin(t) #pandasのdataframe形式にsin(t)を追加してみましょう
df
```

```python
df.plot() #pandasのdataframe形式でもグラフ表示ができます
#x,yを指定しないでグラフ表示をしてみましょう
```

```python
df.plot(x='time') #xを指定するとmatplotlibと同じグラフが得られました
```
